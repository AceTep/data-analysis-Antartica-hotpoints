### 3. Deskriptivna statistika
a. Izvršite deskriptivnu statističku analizu podataka koristeći funkcije Pandas i NumPy biblioteka, poput izračuna srednje vrijednosti, standardne devijacije, koeficijenta varijacije i slično.

```python
import pandas as pd
import numpy as np

# Učitavanje podataka iz CSV datoteke u Pandas DataFrame
df = pd.read_csv('ocisceni_skup_podataka.csv')

# Izdvajanje samo numeričkih stupaca za deskriptivnu statistiku
numeric_df = df.select_dtypes(include=[np.number])

# Izračun srednje vrijednosti za numeričke stupce
mean_values = numeric_df.mean()
print("Srednje vrijednosti:\n", mean_values)

# Izračun standardne devijacije za numeričke stupce
std_deviation = numeric_df.std()
print("\nStandardne devijacije:\n", std_deviation)

# Izračun koeficijenta varijacije za numeričke stupce
coefficient_of_variation = std_deviation / mean_values * 100
print("\nKoeficijenti varijacije:\n", coefficient_of_variation)

# Ostale statističke mjere:
# Izračun minimalnih vrijednosti za numeričke stupce
min_values = numeric_df.min()
print("\nMinimalne vrijednosti:\n", min_values)

# Izračun maksimalnih vrijednosti za numeričke stupce
max_values = numeric_df.max()
print("\nMaksimalne vrijednosti:\n", max_values)

# Izračun kvantila za numeričke stupce
quantiles = numeric_df.quantile([0.25, 0.5, 0.75])
print("\nKvantili (25%, 50%, 75%):\n", quantiles)

# Koristite funkciju describe() za dobivanje sveobuhvatnog pregleda deskriptivne statistike
descriptive_statistics = numeric_df.describe()
print("\nSveobuhvatna deskriptivna statistika:\n", descriptive_statistics)
```

### 4. Inferencijalna statistika
a. Izvršite inferencijalnu statističku analizu podataka koristeći funkcije stats biblioteke (ili slične), poput izračuna p-vrijednosti, konstrukcije intervala pouzdanosti i slično.

```python
import pandas as pd
from scipy.stats import f_oneway
from scipy.stats import ttest_ind
from scipy.stats import t
from scipy.stats import chi2_contingency



df = pd.read_csv('ocisceni_skup_podataka.csv')

grouped = df.groupby('type')

# Spremanje grupa u varijable
group0 = grouped.get_group(0)
group1 = grouped.get_group(1)
group2 = grouped.get_group(2)

# Sada možemo izvršiti inferencijalnu statističku analizu za ove tri grupe podataka
# Na primjer, izračunati ANOVA test ili t-test ovisno o vašem cilju analize

# Izvršavanje ANOVA testa
f_statistic, p_value = f_oneway(group0['brightness'], group1['brightness'], group2['brightness'])

# Ispis rezultata testa
print("ANOVA test za usporedbu srednjih vrijednosti između tri grupe:")
print("F-statistika:", f_statistic)
print("p-vrijednost:", p_value)

# T-test za usporedbu grupa 0 i 1
t_statistic_01, p_value_01 = ttest_ind(group0['brightness'], group1['brightness'])

# Ispis rezultata testa
print("\nT-test za usporedbu srednjih vrijednosti između grupa 0 i 1:")
print("t-statistika:", t_statistic_01)
print("p-vrijednost:", p_value_01)

# T-test za usporedbu grupa 0 i 2
t_statistic_02, p_value_02 = ttest_ind(group0['brightness'], group2['brightness'])

# Ispis rezultata testa
print("\nT-test za usporedbu srednjih vrijednosti između grupa 0 i 2:")
print("t-statistika:", t_statistic_02)
print("p-vrijednost:", p_value_02)

# T-test za usporedbu grupa 1 i 2
t_statistic_12, p_value_12 = ttest_ind(group1['brightness'], group2['brightness'])

# Ispis rezultata testa
print("\nT-test za usporedbu srednjih vrijednosti između grupa 1 i 2:")
print("t-statistika:", t_statistic_12)
print("p-vrijednost:", p_value_12)


from scipy.stats import t

# Zadane razine pouzdanosti
confidence_level = 0.95

# Broj uzoraka u svakoj grupi
n0 = len(group0['brightness'])
n1 = len(group1['brightness'])
n2 = len(group2['brightness'])

# Srednje vrijednosti svake grupe
mean0 = group0['brightness'].mean()
mean1 = group1['brightness'].mean()
mean2 = group2['brightness'].mean()

# Standardne devijacije svake grupe
std0 = group0['brightness'].std()
std1 = group1['brightness'].std()
std2 = group2['brightness'].std()

# Izračun standardne greške srednje vrijednosti svake grupe
sem0 = std0 / (n0 ** 0.5)
sem1 = std1 / (n1 ** 0.5)
sem2 = std2 / (n2 ** 0.5)

# Izračun t-vrijednosti za zadani stupanj slobode (n-1) i razinu pouzdanosti
t_value = t.ppf((1 + confidence_level) / 2, min(n0, n1, n2) - 1)

# Izračun intervala pouzdanosti za srednju vrijednost svake grupe
ci0 = (mean0 - t_value * sem0, mean0 + t_value * sem0)
ci1 = (mean1 - t_value * sem1, mean1 + t_value * sem1)
ci2 = (mean2 - t_value * sem2, mean2 + t_value * sem2)

# Ispis rezultata
print("Interval pouzdanosti za srednju vrijednost grupe 0:", ci0)
print("Interval pouzdanosti za srednju vrijednost grupe 1:", ci1)
print("Interval pouzdanosti za srednju vrijednost grupe 2:", ci2)


# Konstruiramo kontigencijsku tablicu između type i confidence stupaca
contingency_table = pd.crosstab(df['type'], df['confidence'])

```

```python
# Chi-kvadrat test
chi2, p, _, _ = chi2_contingency(contingency_table)
print("\nChi-kvadrat test za provjeru povezanosti između type i confidence:")
print("Chi-kvadrat vrijednost:", chi2)
print("p-vrijednost:", p)

# Pearsonov koeficijent korelacije između brightness i bright_t31
correlation = df['brightness'].corr(df['bright_t31'])
print("\nPearsonov koeficijent korelacije između brightness i bright_t31:", correlation)

```
